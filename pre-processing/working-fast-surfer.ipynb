```python
import nibabel as nib
import numpy as np

seg = nib.load("./output/subject1/mri/aparc.DKTatlas+aseg.deep.mgz")  # or aseg.mgz if DKT not used
seg_data = seg.get_fdata()
voxel_volume = np.prod(seg.header.get_zooms())

left_hip = np.sum(seg_data == 17) * voxel_volume
right_hip = np.sum(seg_data == 53) * voxel_volume

print(f"Left Hippocampus: {left_hip:.2f} mm³")
print(f"Right Hippocampus: {right_hip:.2f} mm³")
```
